```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
original = cv2.imread('/content/chitay.jpg')
img = cv2.cvtColor(original, cv2.COLOR_BGR2GRAY)
plt.imshow(img, "gray")
```

```python
img = cv2.equalizeHist(img)
plt.imshow(img)
```

```python
img = cv2.GaussianBlur(img, (9, 9), 0)
plt.imshow(img)
```

```python
img = cv2.Canny(img, 40, 80)
plt.imshow(img)
```

```python
lined = np.copy(original) * 0
lines = cv2.HoughLinesP(img, 1, np.pi / 180, 15, np.array([]), 50, 20)
for line in lines:
    for x1, y1, x2, y2 in line:
        cv2.line(lined, (x1, y1), (x2, y2), (255, 255, 0))
plt.imshow(lined)
```

```python
output = cv2.addWeighted(original, 0.8, lined, 1, 0)
plt.imshow(output)
```
